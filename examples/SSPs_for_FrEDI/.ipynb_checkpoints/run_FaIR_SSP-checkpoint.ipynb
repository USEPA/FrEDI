## Import Input Data 

### Read in emissions from ssp module

```python
# Get SSP modules (for Fair v1.6)
from fair.SSPs import ssp119, ssp126, ssp245, ssp370, ssp370_lowNTCF, ssp434, ssp460,ssp534, ssp585
```

## Run FaIR

```python
#Run emission scenarios
#solar and volanic forcing variability is turned off
#all settings set to defaults

C_119, F_119, T_119 = fair.forward.fair_scm(emissions=ssp119.Emissions.emissions)#, useMultigas = False)
C_126, F_126, T_126 = fair.forward.fair_scm(emissions=ssp126.Emissions.emissions)
C_245, F_245, T_245 = fair.forward.fair_scm(emissions=ssp245.Emissions.emissions)
C_370, F_370, T_370 = fair.forward.fair_scm(emissions=ssp370.Emissions.emissions)
C_370lowNTCF, F_370lowNTCF, T_370lowNTCF = fair.forward.fair_scm(emissions=ssp370_lowNTCF.Emissions.emissions)
C_434, F_434, T_434 = fair.forward.fair_scm(emissions=ssp434.Emissions.emissions)
C_460, F_460, T_460 = fair.forward.fair_scm(emissions=ssp460.Emissions.emissions)
C_534, F_534, T_534 = fair.forward.fair_scm(emissions=ssp534.Emissions.emissions)
C_585, F_585, T_585 = fair.forward.fair_scm(emissions=ssp585.Emissions.emissions)
```

## Save Data and Format for FrEDI

```python
#save csv file that contains timeseries of global emissions, concentrations, forcing, temperature anomoly

#FrEDI takes temperature change relative to 1995 
# ***QUESTION***Should the FaIR 1986-2005 temperature be averaged? This results in negative temp change
# for both the years 1995 and 1996

#FrEDI year range
start_idx = 1995 - 1765
stop_idx = 2300 - 1765 +1

#scenario output
#data = np.column_stack((emissions_scenario[:,0], T_scenario,np.sum(F_scenario, axis=1),\
#                        emissions_scenario[:,1:],C_scenario))
#df_scenario_out = pd.DataFrame(data,columns = df_cols)
#df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
#print('Full FaIR scenario results')
#display(df_scenario_out)
#df_scenario_out.to_csv(output_dir+'output_scenario_full_FaIR.csv',index=False)

##FrEDI scenario output 
##calc 1986-2005 average
avg = np.mean(T_119[start_idx-10:start_idx+10])
data = np.column_stack((ssp119.Emissions.emissions[start_idx:stop_idx,0], \
                        T_119[start_idx:stop_idx]-avg))
data[data<0] = 0
## OR just calc relative to 1995 value
#data = np.column_stack((ssp119.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_119[start_idx:stop_idx]-T_119[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP1-1.9')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP119_globaltemp_4FrEDI.csv',index=False)

#126
avg = np.mean(T_126[start_idx-10:start_idx+10])
data = np.column_stack((ssp126.Emissions.emissions[start_idx:stop_idx,0], \
                        T_126[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp126.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_126[start_idx:stop_idx]-T_126[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP1-2.6')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP126_globaltemp_4FrEDI.csv',index=False)

#245
avg = np.mean(T_245[start_idx-10:start_idx+10])
data = np.column_stack((ssp245.Emissions.emissions[start_idx:stop_idx,0], \
                        T_245[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp245.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_245[start_idx:stop_idx]-T_245[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP2-4.5')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP245_globaltemp_4FrEDI.csv',index=False)

#370
avg = np.mean(T_370[start_idx-10:start_idx+10])
data = np.column_stack((ssp370.Emissions.emissions[start_idx:stop_idx,0], \
                        T_370[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp370.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_370[start_idx:stop_idx]-T_370[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP3-7.0')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP370_globaltemp_4FrEDI.csv',index=False)

#370_lowNTCF
avg = np.mean(T_370lowNTCF[start_idx-10:start_idx+10])
data = np.column_stack((ssp370_lowNTCF.Emissions.emissions[start_idx:stop_idx,0], \
                        T_370lowNTCF[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp370_lowNTCF.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_370lowNTCF[start_idx:stop_idx]-T_370lowNTCF[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP3-7.0-lowNTCF')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP370_lowNTCF_globaltemp_4FrEDI.csv',index=False)

#434
avg = np.mean(T_434[start_idx-10:start_idx+10])
data = np.column_stack((ssp434.Emissions.emissions[start_idx:stop_idx,0], \
                        T_434[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp434.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_434[start_idx:stop_idx]-T_434[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP4-3.4')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP434_globaltemp_4FrEDI.csv',index=False)

#460
avg = np.mean(T_460[start_idx-10:start_idx+10])
data = np.column_stack((ssp460.Emissions.emissions[start_idx:stop_idx,0], \
                        T_460[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp460.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_460[start_idx:stop_idx]-T_460[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP4-6.0')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP460_globaltemp_4FrEDI.csv',index=False)

#534
avg = np.mean(T_534[start_idx-10:start_idx+10])
data = np.column_stack((ssp534.Emissions.emissions[start_idx:stop_idx,0], \
                        T_534[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp534.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_534[start_idx:stop_idx]-T_534[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP5-3.4')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP534_globaltemp_4FrEDI.csv',index=False)

#585
avg = np.mean(T_585[start_idx-10:start_idx+10])
data = np.column_stack((ssp585.Emissions.emissions[start_idx:stop_idx,0], \
                        T_585[start_idx:stop_idx]-avg))
data[data<0] = 0
#data = np.column_stack((ssp585.Emissions.emissions[start_idx:stop_idx,0], \
#                        T_585[start_idx:stop_idx]-T_585[start_idx]))
df_scenario_out = pd.DataFrame(data,columns = ['Year','temp_c'])
df_scenario_out.loc[:,'Year'] = df_scenario_out.loc[:,'Year'].astype(int)
print('SSP5-8.5')
display(df_scenario_out)
df_scenario_out.to_csv(output_dir+'SSP585_globaltemp_4FrEDI.csv',index=False)
```
